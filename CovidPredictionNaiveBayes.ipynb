```python
import pandas as pd

def probability(list,n):
    prob= []
    for i in range(len(list)):
        probability= list[i]/n
        prob.append(probability)
    return prob
    
dataset= pd.read_csv("./covid data.csv")

cough= dataset["Sore Throat and Cough"]
runnose= dataset["Runny Nose"]
fever= dataset["Fever"]
diffbreath= dataset["Difficulty in Breathing"]
covid= dataset["COVID-19?"]

n= len(covid)

covidposcough, covidnegcough= [],[]
covidposrunnose, covidnegrunnose= [],[]
covidposfever, covidnegfever= [],[]
covidposdiffbreath, covidnegdiffbreath= [],[]
covidcnt= []

count1, count2= [0,0], [0,0]
for i in range(len(cough)):
        if cough[i]=="Yes":
            if covid[i]=="Yes":
                count1[0]+=1
            else:
                count2[0]+=1
        else:
            if covid[i]=="Yes":
                count1[1]+=1
            else:
                count2[1]+=1
for i in range(len(count1)):
    covidposcough.append(count1[i])
    covidnegcough.append(count2[i])

count1, count2= [0,0], [0,0]
for i in range(len(runnose)):
        if runnose[i]=="Yes":
            if covid[i]=="Yes":
                count1[0]+=1
            else:
                count2[0]+=1
        else:
            if covid[i]=="Yes":
                count1[1]+=1
            else:
                count2[1]+=1
for i in range(len(count1)):
    covidposrunnose.append(count1[i])
    covidnegrunnose.append(count2[i])

count1, count2= [0,0], [0,0]
for i in range(len(fever)):
        if fever[i]=="Yes":
            if covid[i]=="Yes":
                count1[0]+=1
            else:
                count2[0]+=1
        else:
            if covid[i]=="Yes":
                count1[1]+=1
            else:
                count2[1]+=1
for i in range(len(count1)):
    covidposfever.append(count1[i])
    covidnegfever.append(count2[i])

count1, count2= [0,0,0], [0,0,0]
for i in range(len(diffbreath)):
        if diffbreath[i]=="Strong":
            if covid[i]=="Yes":
                count1[0]+=1
            else:
                count2[0]+=1
        elif diffbreath[i]=="Mild":
            if covid[i]=="Yes":
                count1[1]+=1
            else:
                count2[1]+=1
        else:
            if covid[i]=="Yes":
                count1[2]+=1
            else:
                count2[2]+=1
for i in range(len(count1)):
    covidposdiffbreath.append(count1[i])
    covidnegdiffbreath.append(count2[i])

count= [0,0]
for i in range(len(covid)):
        if covid[i]=="Yes":
            count[0]+=1
        else:
            count[1]+=1
for i in range(len(count)):
    covidcnt.append(count[i])
    
covidprob= probability(covidcnt,n)
covposcoughprob= probability(covidposcough, covidcnt[0])
covnegcoughprob= probability(covidnegcough, covidcnt[1])
covposrunnoseprob= probability(covidposrunnose, covidcnt[0])
covnegrunnoseprob= probability(covidnegrunnose, covidcnt[1])
covposfeverprob= probability(covidposfever, covidcnt[0])
covnegfeverprob= probability(covidnegfever, covidcnt[1])
covposdiffbreathprob= probability(covidposdiffbreath, covidcnt[0])
covnegdiffbreathprob= probability(covidnegdiffbreath, covidcnt[1])

scough= input("Do you have sore throat and cough?(Yes/No) ").upper()
srunnose= input("Do you have runny nose?(Yes/No) ").upper()
sfever= input("Do you have fever?(Yes/No) ").upper()
sdiffbreath= input("Do you have difficulty in breathing?(Strong/Mild/No) ").upper()

if scough=="YES":
    val1= covposcoughprob[0]
    val2= covnegcoughprob[0]
elif scough=="NO":
    val1= covposcoughprob[1]
    val2= covnegcoughprob[1]
else:
    print("Error in entering symptoms.")
    
if srunnose=="YES":
    val3= covposrunnoseprob[0]
    val4= covnegrunnoseprob[0]
elif srunnose=="NO":
    val3= covposrunnoseprob[1]
    val4= covnegrunnoseprob[1]
else:
    print("Error in entering symptoms.")
    
if sfever=="YES":
    val5= covposfeverprob[0]
    val6= covnegfeverprob[0]
elif sfever=="NO":
    val5= covposfeverprob[1]
    val6= covnegfeverprob[1]
else:
    print("Error in entering symptoms.")
    
if sdiffbreath=="STRONG":
    val7= covposdiffbreathprob[0]
    val8= covnegdiffbreathprob[0]
elif sdiffbreath=="MILD":
    val7= covposdiffbreathprob[1]
    val8= covnegdiffbreathprob[1]
elif sdiffbreath=="NO":
    val7= covposdiffbreathprob[2]
    val8= covnegdiffbreathprob[2]
else:
    print("Error in entering symptoms.")
    
finalval1= val1*val3*val5*val7*covidprob[0]
finalval2= val2*val4*val6*val8*covidprob[1]

if finalval1>finalval2:
    print("\nResult:\nCOVID POSITIVE")
else:
    print("\nResult:\nCOVID NEGATIVE")
```
